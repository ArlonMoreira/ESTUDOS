```python
#Importando a base de dados
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt  

#Abrundo o csv
df = pd.read_csv(r'C:\Users\Arlon\Desktop\ESTUDOS\Arquivos\pima-data.csv', sep=',', names=['num_preg', 'glucose_conc', 'diastolic_bp', 'thickness', 'insulin', 'bmi', 'diab_pred', 'age', 'diabetes'])

#Verificando o tamanho do dataframe
df.head(5)
```

```python
#verificando a existência de relação entre as variáveis.
# +1  = forte correlação positiva // Ou seja, valores diretamente proporcionais
# 0   = não há correlação // Ausência de correção
# -1  = forte correlação negativa // Inversamente proporcionais

def plot_corr(df, size=10):
    corr = df.corr()
    fig, ax = plt.subplots(figsize = (size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns) 
    plt.yticks(range(len(corr.columns)), corr.columns)
    
plot_corr(df)
#Quanto mais proximo do amarelo mais forte será a correção existentes entre as variáveis
```

```python
#Outra forma de verificar a correção
df.corr()
```

```python
#Parte importante, definir as classses
diabetes_map = {True : 1, False : 0}
```

```python
#A função Map aplica uma função a todos os elementos de um objeto, seja uma lista, um dataframe. Neste caso fará uma substituição
#de True para 1 e False para 0

# Aplicando o mapeamento ao dataset
df['diabetes'] = df['diabetes'].map(diabetes_map)

df.head(5)
```
